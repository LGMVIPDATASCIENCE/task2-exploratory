### Cleaning the data

```python
a=a[["iyear","imonth","iday","country_txt","region_txt","provstate","city",
     "latitude","longitude","location","summary","attacktype1_txt","targtype1_txt",
     "gname","motive","weaptype1_txt","nkill","nwound","addnotes"]]
```

```python
a.head()
```

```python
a.rename(columns={'iyear':'Year',"imonth":'Month',"iday":"Day","country_txt":"Country","region_txt":"Region","provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude","location":"Location","summary":"Summary","attacktype1_txt":"Attack Type","targtype1_txt":"Target Type","gname":"Group Name","motive":"Motive","weaptype1_txt":"Weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add Notes"},inplace=True)
```

```python
a.head()
```

```python
a.info()
```

```python
a.shape
```

```python
a.isnull().sum()
```

```python
a["Killed"]=a["Killed"].fillna(0)
a["Wounded"]=a["Wounded"].fillna(0)
a["Casualty"]=a["Killed"]+a["Wounded"]
```

```python
a.describe()
```

### Observation
 * The data consists of terrorist activities ranging from       the year: 1970 to 2017
 * Maximum number of people killed in an event were: 1570
 * Maximum number of people wounded in an event were:8191
 * Maximum number of total casualties in an even  
    were:9474

### Visualizing the data

#### 1.Year wise Attacks

* Number of Attacks in each Year

```python
attacks=a["Year"].value_counts(dropna=False).sort_index().to_frame().reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.head()
```

```python
attacks.plot(kind="bar",color="blue",figsize=(15,6),fontsize=13)
plt.title("Timeline of Attacks",fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks",fontsize=15)
plt.show()
```

* (i). Most number of attacks(16903) in 2014
* (ii). Least number of attacks(471) in 1971









* Total Casualties (Killed + Wounded) in each Year

```python
y=a[["Year","Casualty"]].groupby("Year").sum()
y.head()
```

```python
y.plot(kind="bar",color="blue",figsize=(15,6))
plt.title("Year wise Casualties",fontsize=13)
plt.xlabel("Years",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

* Killed in each Year

```python
y1=a[["Year","Wounded"]].groupby("Year").sum()
y1.head()
```

* Wounded in each Region

```python
y2=a[["Year","Wounded"]].groupby("Year").sum()
y2.head()
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,1,1)
ax1=fig.add_subplot(2,1,2)

#killed
y1.plot(kind="bar",color="blue",figsize=(15,15),ax=ax0)
ax0.set_title("People Killed in each Year")
ax0.set_xlabel("Years")
ax0.set_ylabel("Number of People Killed")

#wounded
y2.plot(kind="bar",color="blue",figsize=(15,15),ax=ax1)
ax1.set_title("People Wounded in each Year")
ax1.set_xlabel("Years")
ax1.set_ylabel("Number of People Wounded")

plt.show() 
```

### 2.Region wise Attacks

* Distributiion of Terrorist Attacks over Regions from 1970-2017

```python
reg=pd.crosstab(a.Year,a.Region)
reg.head()
```

```python
reg.plot(kind="area", stacked=False, alpha=0.5,figsize=(20,10))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

* Total Terrorist Attacks in each Region from 1970-2017

```python
regt=reg.transpose()
regt["Total"]=regt.sum(axis=1)
ra=regt["Total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar",figsize=(15,6))
plt.title("Total Number of Attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("Number of Attacks")
plt.show()
```

* Total Casualties (Killed + Wounded) in each Region

```python
rc=a[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending=False)
rc
```

```python
rc.plot(kind="bar",color="blue",figsize=(15,6))
plt.title("Region wise Casualties",fontsize=13)
plt.xlabel("Regions",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()
```

* Killed in each Region

```python
rk=a[["Region","Killed"]].groupby("Region").sum().sort_values(by="Killed",ascending=False)
rk
```

* Wounded in each Region

```python
rw=a[["Region","Wounded"]].groupby("Region").sum().sort_values(by="Wounded",ascending=False)
rw
```

```python
fig=plt.figure()
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#Killed
rk.plot(kind="bar",color="blue",figsize=(15,6),ax=ax0)
ax0.set_title("People Killed in each Region")
ax0.set_xlabel("Regions")
ax0.set_ylabel("Number of People Killed")

#Wounded
rw.plot(kind="bar",color="blue",figsize=(15,6),ax=ax1)
ax1.set_title("People Wounded in each Region")
ax1.set_xlabel("Regions")
ax1.set_ylabel("Number of People Wounded")
plt.show()
```
